```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("secret_data.csv")
```

```python
# data = data.iloc[:999]
```

```python
data
```

```python
# убираем среднее
for col in data.columns:
    data[col]-=data[col].mean()
```

```python
data=data.T
data.head()
```

```python
CX = 1/999*(data.dot(data.T))
```

```python
CX
```

```python
eig = np.linalg.eig(CX)
```

```python
eig
```

```python
P = pd.DataFrame(data=eig[1], index = CX.columns, columns=CX.columns)
```

```python
P
```

```python
CY = P.dot(CX).dot(P.T)
```

```python
CY
```

```python
# всё, что не по диагонали - нули (ну пачти)
```

```python
d = pd.Series(np.diag(CY), index= CY.index)
```

```python
d # это дисперсии всех фичей - их важность: чем она меньше, тем лучше её дропнуть
```

```python
d.sort_values(ascending=False, inplace=True)
```

```python
d
```

```python
d.plot()
```

```python
d.head(250).plot()
```

```python
d.head(300).plot()
```

```python
# го оставим 300 (лол)
```

```python
to_kill = d[300:].index
```

```python
for col in to_kill:
    data.drop(col, axis=0, inplace=True)
```

```python
data=data.T
```

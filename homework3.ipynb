### Первая задача

```python
def fibonacci():
    a, b = 1, 1
    while (True):
        if (b>a):
            a,b=b,a
        b=a+b
        yield b
```

```python
print("Enter the limit!")
n=int(input())
print ("Fibonacci numbers that are <=", n, ':')
itr= fibonacci()
if (n>=1):
    print (1)
    print (1)
    t=next(itr)
    while (t<=n):
        print(t)
        t=next(itr)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
def get_boxplot(train_df,param):
    sns.boxplot(y=train_df[param])
    plt.title(param)
    plt.show()
    mean_val = train_df[param].mean(skipna=True)
    quantile_25 = train_df[param].quantile(0.25)
    quantile_75 = train_df[param].quantile(0.75)
    IQR = quantile_75 - quantile_25
    min_lim = quantile_25 - 1.5*IQR
    if min_lim < 0:
        min_lim = 0
    max_lim = quantile_75 + 1.5*IQR

    print("{:<15} : {:<15}".format("Mean",str(mean_val)))
    print("{:<15} : {:<15}".format("25th Quantile",str(quantile_25)))
    print("{:<15} : {:<15}".format("75th Quantile",str(quantile_75)))
    print("{:<15} : {:<15}".format("Minimum",str(min_lim)))
    print("{:<15} : {:<15}".format("Maximum",str(max_lim)))
    return mean_val,min_lim,max_lim
```

```python
mean_age,min_lim_age,max_lim_age = get_boxplot(train_main,"Age")
```

```python
import torch
import torchvision
import torchvision.datasets as datasets
from torch.utils.data import DataLoader
import torch.utils.data as data_utils
from os.path import exists
import matplotlib.pyplot as plt
import numpy as np
lr = 0.01 #скорость обучения
batch_size = 512
dtype = torch.float
device = torch.device('cuda') if torch.cuda.is_available() else torch.device("cpu")
```

```python
def loss_plot(fig, ax, train_loss, test_loss, loss_name):
    train_line = ax.plot(train_loss, color = 'black')
    test_line = ax.plot(test_loss, color = 'red')
    #рисуем вертикальные линии с эпохами
    i = 1
    while i * len(train_loader) < len(train_loss):
        ax.axvline(i * len(train_loader), linestyle = '--')
        i += 1
    ax.set_xlabel('Батч')
    ax.set_ylabel(loss_name)
    ax.legend(('Тренировочная выборка', 'Тестовая выборка'))
    fig.canvas.draw()
```

```python
y_train = torch.tensor(y_train.values.astype(np.float32))
x_train = torch.tensor(x_train.values.astype(np.float32))

y_test = torch.tensor(y_test.values.astype(np.float32))
x_test = torch.tensor(x_test.values.astype(np.float32))
```

```python
x_train = x_train.float().to(device)
x_test = x_test.float().to(device)
y_train = y_train.to(device)
y_test = y_test.to(device)
```

```python
x_train.size()
```

```python
# Неоднозначно!!

x_train = x_train.reshape(-1, 10)/128
x_test = x_test.reshape(-1, 10)/128
# x_train = x_train.reshape(-1)
# x_test = x_test.reshape(-1)
y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)
```

```python
train_loader = DataLoader(list(zip(x_train, y_train)), batch_size=batch_size,shuffle=True)
test_loader = DataLoader(list(zip(x_test, y_test)), batch_size=batch_size,shuffle=True)
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(10, 10),
    torch.nn.ReLU(),
    torch.nn.Dropout(p = 0.5),
    torch.nn.Linear(10, 1),
).to(device)
#нужно предварительно создать оптимайзер на параметры
#оптимайзер - это штука, которая на основе градиентов каким-то образом обновляет параметры
optimizer = torch.optim.SGD(model.parameters(), lr = lr)
#кросс-энтропия сама делает softmax для y_hat!(ну это только в торче, внимательнее)
lossf = torch.nn.BCEWithLogitsLoss()
```

```python
y_test = y_test.float()
y_train = y_train.float()
```

```python
check=[]
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(8,10))
ax = fig.subplots(2,1)
fig.tight_layout(pad=3.0)

ax1, ax2 = ax
train_loss = []
test_loss = []
train_acc = []
test_acc = []
epochs =10
for i in range(epochs):
    for x, y in train_loader:
        y_hat = model(x)
        
        y_hat=y_hat.float()
        check.append(y_hat)
        loss = lossf(y_hat.reshape(-1), y)
        train_loss.append(loss.item())
        train_acc.append(torch.mean((y_train==torch.round(abs(y_hat))).float()))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        with torch.no_grad():
            y_hat = model(x_test)
            test_loss.append(lossf(y_hat.reshape(-1), y_test))
            test_acc.append(torch.mean((y_test==torch.round(abs(y_hat))).float()))
    
    fig.suptitle('Сеть с одним скрытым слоем', fontsize=16)
    loss_plot(fig,ax1,train_loss, test_loss, "Бинарная кросс-энтропия")
    loss_plot(fig,ax2,train_acc, test_acc, "Точность")
```

```python
model.eval()
test_main = torch.tensor(test_main.values.astype(np.float32))
x_test_main = test_main.reshape(-1, 10)/128
y_hat = model(x_test_main)
```

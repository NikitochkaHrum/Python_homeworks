```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("secret_data.csv")
```

```python
# data = data.iloc[:999]
```

```python
data
```

```python
# убираем среднее
for col in data.columns:
    data[col]-=data[col].mean()
```

```python
data=data.T
data.head()
```

```python
CX = 1/999*(data.dot(data.T))
```

```python
CX
```

```python
eig = np.linalg.eig(CX)
```

```python
eig
```

```python
sver = pd.Series(data=eig[0], index=CY.index)
```

```python
sver.sort_values(ascending=False, inplace=True)
```

```python
sver.plot()
```

```python
sver[:10]
```

```python
to_kill = sver[3:].index
```

```python
for col in to_kill:
    data.drop(col, axis=0, inplace=True)
```

```python
data=data.T
```
